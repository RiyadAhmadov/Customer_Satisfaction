```python
from textblob import TextBlob
from textblob import exceptions
from googletrans import Translator  # Make sure to install googletrans==4.0.0-rc1
```

```python
# Function to translate text to English using Google Translate
def translate_to_english(text):
    translator = Translator()
    try:
        translation = translator.translate(text, dest='en')
        return translation.text
    except exceptions.GoogleTranslateError as e:
        print(f"Translation error: {e}")
        return text  # Return original text in case of an error
```

```python
# Assuming df is your DataFrame and 'column1' is the column containing text
feedback['column1_english'] = feedback['Col1'].apply(translate_to_english)
```

```python
# Function to perform sentiment analysis using TextBlob
def analyze_sentiment(text):
    blob = TextBlob(text)
    return blob.sentiment.polarity  # You can also use sentiment.subjectivity if needed
feedback['sentiment_polarity'] = feedback['column1_english'].apply(analyze_sentiment)
```

```python
feedback['Impressure'] = feedback['sentiment_polarity'].apply(lambda x: 'Negative' if x < 0 else ('Neutral' if x == 0 else 'Positive'))
```
